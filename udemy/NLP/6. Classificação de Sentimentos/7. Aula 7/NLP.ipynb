Pré-Processamento

```python
pontuacoes = string.punctuation
```

```python
pontuacoes
```

```python
from spacy.lang.pt.stop_words import STOP_WORDS
```

```python
nlp = spacy.load("pt")
```

```python
def pre_processing(texto):
  texto = texto.lower()
  doc = nlp(texto)
  lista = []

  for token in doc:
    lista.append(token.lemma_)
  
  lista = [palavra for palavra in lista if palavra not in STOP_WORDS and palavra not in pontuacoes] # tirar pontuações e stop_words
  lista = [token for token in lista if not token.isdigit()] # tirar digitos
  string_lista = " ".join(lista) # unir em uma string esse array lista criado

  return string_lista
```

```python
teste = "2 3 10 String de teste para identificar se a funcao 40 pre_processing está indo tudo bem aqui em Fortaleza e vou colocar pontuacoes, . ; para testar!! haha,,,...& 555"
```

```python
pre_processing(teste)
```

Limpeza dos textos da base de dados

```python
base_dados["texto"] = base_dados["texto"].apply(pre_processing)
```

```python
base_dados.head()
```

```python
base_dados_final = []
for texto, emocao in zip(base_dados['texto'], base_dados['emocao']):
  if emocao == "alegria":
    dic = {"Alegria": True, "Medo": False}
  elif emocao == "medo":
    dic = {"Alegria": False, "Medo": True}
  base_dados_final.append([texto, dic.copy()])
```

```python
base_dados_final[0]
```

Criação do Classificador

```python
modelo = spacy.blank('pt')
categorias = modelo.create_pipe("textcat")
categorias.add_label("Alegria")
categorias.add_label("Medo")
modelo.add_pipe(categorias)
historico = []
```

```python
modelo.begin_training()
for epoca in range(1000):
  random.shuffle(base_dados_final)
  losses = {}
  for batch in spacy.util.minibatch(base_dados_final, 30):
    textos = [modelo(texto) for texto, entities in batch]
    annotations = [{"cats": entities} for texto, entities in batch]
    modelo.update(textos, annotations, losses = losses)
  if epoca % 100 == 0:
    print(losses)
    historico.append(losses)
```

```python
historico_loss = []

for token in historico:
  historico_loss.append(token.get("textcat"))

historico_loss = np.array(historico_loss)
```

```python
plt.plot(historico_loss)
plt.title('Progressao do Erro')
```

```python
modelo.to_disk("modelo_pronto")
```
